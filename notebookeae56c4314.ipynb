```python
import pandas as pd
import os

csv_path = "/kaggle/input/datasets/altiarikaeka/features-all-10methods/Features_All_10Methods(2).csv"

df_features = pd.read_csv(csv_path)

print("Dataset berhasil diload.")
print("Shape dataset:", df_features.shape)

print("\nNama kolom:")
print(df_features.columns.tolist())

print("\nPreview data:")
display(df_features.head())
```

```python
required_cols = ["Label", "Condition", "subject_file"]

for col in required_cols:
    if col in df_features.columns:
        print(f"{col}: ADA")
    else:
        print(f"{col}: TIDAK ADA")

print("\nDistribusi Label:")
print(df_features["Label"].value_counts())

print("\nDistribusi Condition:")
print(df_features["Condition"].value_counts())
```

```python
# ============================================================
# TAHAP 1: Filter Gamma & Persiapan Data
# ============================================================

# 1. Definisikan 8 kolom Gamma
gamma_cols = [
    'RAW_TP9_BP_Gamma',  'RAW_TP9_RP_Gamma',
    'RAW_AF7_BP_Gamma',  'RAW_AF7_RP_Gamma',
    'RAW_AF8_BP_Gamma',  'RAW_AF8_RP_Gamma',
    'RAW_TP10_BP_Gamma', 'RAW_TP10_RP_Gamma'
]

# 2. Kolom metadata (bukan fitur, tidak masuk model)
metadata_cols = ['Epoch', 'subject_file', 'Condition', 'Label', 'epoch_id']

# 3. Pastikan semua kolom ada di dataframe
missing_gamma = [c for c in gamma_cols if c not in df_features.columns]
missing_meta  = [c for c in metadata_cols if c not in df_features.columns]

if missing_gamma:
    print(f"PERINGATAN: Kolom Gamma tidak ditemukan: {missing_gamma}")
else:
    print("Semua 8 kolom Gamma ditemukan.")

if missing_meta:
    print(f"PERINGATAN: Kolom metadata tidak ditemukan: {missing_meta}")
else:
    print("Semua kolom metadata ditemukan.")

# 4. Pisahkan X, y, dan metadata
X_gamma = df_features[gamma_cols].copy()
y       = df_features['Label'].copy()
metadata = df_features[['Epoch', 'subject_file', 'Condition', 'Label']].copy()

# 5. Verifikasi hasil
print(f"\nShape X (fitur Gamma) : {X_gamma.shape}")
print(f"Shape y (label)       : {y.shape}")
print(f"Shape metadata         : {metadata.shape}")

print(f"\nDistribusi label:")
print(y.value_counts().sort_index())

print(f"\nDistribusi kondisi:")
print(metadata['Condition'].value_counts().sort_index())

print(f"\nPreview X_gamma (5 baris pertama):")
display(X_gamma.head())

print(f"\nStatistik deskriptif X_gamma:")
display(X_gamma.describe())
```

```python
# ============================================================
# TAHAP 2: Train-Validation-Test Split + Scaling
# ============================================================

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np

# ---- 1. Train/Val/Test Split (70/15/15) dengan stratify ----
# Pertama: pisahkan test set (15%)
X_trainval, X_test, y_trainval, y_test, meta_trainval, meta_test = train_test_split(
    X_gamma, y, metadata,
    test_size=0.15,
    random_state=42,
    stratify=y
)

# Kedua: pisahkan train dan validation dari sisa (85%)
# 15% dari 85% вүҲ 17.6% dari total, sehingga train вүҲ 70%, val вүҲ 15%, test вүҲ 15%
X_train, X_val, y_train, y_val, meta_train, meta_val = train_test_split(
    X_trainval, y_trainval, meta_trainval,
    test_size=0.176,  # 0.176 dari 85% вүҲ 15% dari total
    random_state=42,
    stratify=y_trainval
)

print("=== Ukuran Split ===")
print(f"Train      : {X_train.shape[0]} sampel ({X_train.shape[0]/len(X_gamma)*100:.1f}%)")
print(f"Validation : {X_val.shape[0]} sampel ({X_val.shape[0]/len(X_gamma)*100:.1f}%)")
print(f"Test       : {X_test.shape[0]} sampel ({X_test.shape[0]/len(X_gamma)*100:.1f}%)")
print(f"Total      : {X_train.shape[0] + X_val.shape[0] + X_test.shape[0]} sampel")

# ---- 2. Cek distribusi label di setiap split ----
print("\n=== Distribusi Label per Split ===")
for name, ys in [("Train", y_train), ("Validation", y_val), ("Test", y_test)]:
    dist = ys.value_counts().sort_index()
    print(f"{name:12s} -> Label 0: {dist[0]:5d}, Label 1: {dist[1]:5d}, "
          f"Ratio 0:1 = {dist[0]/dist[1]:.2f}")

# ---- 3. Scaling (fit hanya pada training data!) ----
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)   # fit + transform
X_val_scaled   = scaler.transform(X_val)          # transform saja
X_test_scaled  = scaler.transform(X_test)         # transform saja

print("\n=== Setelah Scaling (statistik X_train) ===")
print(f"Mean (seharusnya ~0): {X_train_scaled.mean(axis=0).round(4)}")
print(f"Std  (seharusnya ~1): {X_train_scaled.std(axis=0).round(4)}")

# ---- 4. Reshape untuk CNN: (samples, 8, 1) ----
X_train_cnn = X_train_scaled.reshape(-1, 8, 1)
X_val_cnn   = X_val_scaled.reshape(-1, 8, 1)
X_test_cnn  = X_test_scaled.reshape(-1, 8, 1)

print(f"\n=== Shape untuk CNN ===")
print(f"X_train_cnn : {X_train_cnn.shape}")
print(f"X_val_cnn   : {X_val_cnn.shape}")
print(f"X_test_cnn  : {X_test_cnn.shape}")

# ---- 5. Simpan juga versi unscaled untuk RF (Random Forest tidak butuh scaling) ----
X_train_unscaled = X_train.values
X_val_unscaled   = X_val.values
X_test_unscaled  = X_test.values

print("\nTahap 2 selesai. Variabel yang tersedia:")
print("  - X_train_scaled, X_val_scaled, X_test_scaled  (untuk SVM)")
print("  - X_train_cnn, X_val_cnn, X_test_cnn          (untuk CNN)")
print("  - X_train_unscaled, X_val_unscaled, X_test_unscaled (untuk RF)")
print("  - y_train, y_val, y_test                       (target)")
print("  - meta_train, meta_val, meta_test              (metadata)")
print("  - scaler (StandardScaler yang sudah di-fit)")
```

```python
# ============================================================
# TAHAP 3A: Baseline SVM (Support Vector Machine)
# ============================================================

from sklearn.svm import SVC
from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                             f1_score, confusion_matrix, classification_report)
import time

# ---- 1. Training ----
print("Training SVM (RBF kernel)...")
start = time.time()

svm_model = SVC(kernel='rbf', C=1.0, gamma='scale', random_state=42)
svm_model.fit(X_train_scaled, y_train)

train_time = time.time() - start
print(f"Training selesai dalam {train_time:.2f} detik")

# ---- 2. Prediksi ----
y_pred_val_svm  = svm_model.predict(X_val_scaled)
y_pred_test_svm = svm_model.predict(X_test_scaled)

# ---- 3. Evaluasi Validation Set ----
print("\n" + "="*50)
print("HASIL SVM - VALIDATION SET")
print("="*50)
print(f"Accuracy  : {accuracy_score(y_val, y_pred_val_svm):.4f}")
print(f"Precision : {precision_score(y_val, y_pred_val_svm, average='weighted'):.4f}")
print(f"Recall    : {recall_score(y_val, y_pred_val_svm, average='weighted'):.4f}")
print(f"F1-Score  : {f1_score(y_val, y_pred_val_svm, average='weighted'):.4f}")
print(f"\nConfusion Matrix:\n{confusion_matrix(y_val, y_pred_val_svm)}")
print(f"\n{classification_report(y_val, y_pred_val_svm, target_names=['Label 0 (Before)', 'Label 1 (After)'])}")

# ---- 4. Evaluasi Test Set ----
print("="*50)
print("HASIL SVM - TEST SET")
print("="*50)
print(f"Accuracy  : {accuracy_score(y_test, y_pred_test_svm):.4f}")
print(f"Precision : {precision_score(y_test, y_pred_test_svm, average='weighted'):.4f}")
print(f"Recall    : {recall_score(y_test, y_pred_test_svm, average='weighted'):.4f}")
print(f"F1-Score  : {f1_score(y_test, y_pred_test_svm, average='weighted'):.4f}")
print(f"\nConfusion Matrix:\n{confusion_matrix(y_test, y_pred_test_svm)}")
print(f"\n{classification_report(y_test, y_pred_test_svm, target_names=['Label 0 (Before)', 'Label 1 (After)'])}")
```

```python
# ============================================================
# TAHAP 3C: Baseline CNN
# ============================================================

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (Conv1D, MaxPooling1D, Flatten, Dense,
                                     Dropout, BatchNormalization, Input)
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                             f1_score, confusion_matrix, classification_report)
import time

# ---- 1. Bangun Model CNN ----
model_cnn = Sequential([
    Input(shape=(8, 1)),
    Conv1D(filters=32, kernel_size=3, activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling1D(pool_size=2),

    Conv1D(filters=64, kernel_size=3, activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling1D(pool_size=2),

    Flatten(),
    Dense(64, activation='relu'),
    Dropout(0.5),
    Dense(32, activation='relu'),
    Dropout(0.3),
    Dense(1, activation='sigmoid')
])

model_cnn.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model_cnn.summary()

# ---- 2. Callbacks ----
early_stop = EarlyStopping(
    monitor='val_loss',
    patience=15,
    restore_best_weights=True,
    verbose=1
)

reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=5,
    min_lr=1e-6,
    verbose=1
)

# ---- 3. Training ----
print("\nTraining CNN...")
start = time.time()

history = model_cnn.fit(
    X_train_cnn, y_train,
    validation_data=(X_val_cnn, y_val),
    epochs=100,
    batch_size=32,
    callbacks=[early_stop, reduce_lr],
    verbose=1
)

train_time = time.time() - start
print(f"\nTraining selesai dalam {train_time:.2f} detik")
print(f"Total epoch berjalan: {len(history.history['loss'])}")

# ---- 4. Prediksi ----
y_prob_val_cnn  = model_cnn.predict(X_val_cnn, verbose=0).flatten()
y_prob_test_cnn = model_cnn.predict(X_test_cnn, verbose=0).flatten()

y_pred_val_cnn  = (y_prob_val_cnn >= 0.5).astype(int)
y_pred_test_cnn = (y_prob_test_cnn >= 0.5).astype(int)

# ---- 5. Evaluasi Validation Set ----
print("\n" + "="*50)
print("HASIL CNN - VALIDATION SET")
print("="*50)
print(f"Accuracy  : {accuracy_score(y_val, y_pred_val_cnn):.4f}")
print(f"Precision : {precision_score(y_val, y_pred_val_cnn, average='weighted'):.4f}")
print(f"Recall    : {recall_score(y_val, y_pred_val_cnn, average='weighted'):.4f}")
print(f"F1-Score  : {f1_score(y_val, y_pred_val_cnn, average='weighted'):.4f}")
print(f"\nConfusion Matrix:\n{confusion_matrix(y_val, y_pred_val_cnn)}")
print(f"\n{classification_report(y_val, y_pred_val_cnn, target_names=['Label 0 (Before)', 'Label 1 (After)'])}")

# ---- 6. Evaluasi Test Set ----
print("="*50)
print("HASIL CNN - TEST SET")
print("="*50)
print(f"Accuracy  : {accuracy_score(y_test, y_pred_test_cnn):.4f}")
print(f"Precision : {precision_score(y_test, y_pred_test_cnn, average='weighted'):.4f}")
print(f"Recall    : {recall_score(y_test, y_pred_test_cnn, average='weighted'):.4f}")
print(f"F1-Score  : {f1_score(y_test, y_pred_test_cnn, average='weighted'):.4f}")
print(f"\nConfusion Matrix:\n{confusion_matrix(y_test, y_pred_test_cnn)}")
print(f"\n{classification_report(y_test, y_pred_test_cnn, target_names=['Label 0 (Before)', 'Label 1 (After)'])}")
```

```python
# ============================================================
# TAHAP 4: ACGAN AUGMENTATION UNTUK GAMMA-ONLY FEATURES
# ============================================================

import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, LeakyReLU, BatchNormalization, Dropout, Embedding, Flatten, Concatenate
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from sklearn.utils import shuffle
import matplotlib.pyplot as plt

# ============================================================
# 1. SETUP AWAL
# ============================================================

np.random.seed(42)
tf.random.set_seed(42)

# Data training yang dipakai ACGAN
X_train_acgan = np.array(X_train_scaled, dtype=np.float32)
y_train_acgan = np.array(y_train).astype(int)

feature_dim = X_train_acgan.shape[1]   # 8 fitur Gamma
num_classes = 2                        # Label 0 dan Label 1
noise_dim = 32                         # Dimensi noise generator

print("=== Data untuk ACGAN ===")
print(f"X_train_acgan shape : {X_train_acgan.shape}")
print(f"y_train_acgan shape : {y_train_acgan.shape}")
print(f"Feature dim         : {feature_dim}")
print(f"Jumlah kelas        : {num_classes}")
print(f"Noise dim           : {noise_dim}")

unique, counts = np.unique(y_train_acgan, return_counts=True)
print("\nDistribusi label training asli:")
for u, c in zip(unique, counts):
    print(f"Label {u}: {c} sampel")

# ============================================================
# 2. BUILD GENERATOR
# Generator menerima noise + label, lalu menghasilkan 8 fitur Gamma sintetis
# ============================================================

def build_generator(noise_dim, num_classes, feature_dim):
    noise_input = Input(shape=(noise_dim,), name="noise_input")
    label_input = Input(shape=(1,), dtype="int32", name="label_input")

    label_embedding = Embedding(num_classes, noise_dim)(label_input)
    label_embedding = Flatten()(label_embedding)

    x = Concatenate()([noise_input, label_embedding])

    x = Dense(64)(x)
    x = LeakyReLU(negative_slope=0.2)(x)
    x = BatchNormalization()(x)

    x = Dense(128)(x)
    x = LeakyReLU(negative_slope=0.2)(x)
    x = BatchNormalization()(x)

    x = Dense(64)(x)
    x = LeakyReLU(negative_slope=0.2)(x)

    output = Dense(feature_dim, activation="linear", name="generated_features")(x)

    model = Model([noise_input, label_input], output, name="Generator")
    return model

# ============================================================
# 3. BUILD DISCRIMINATOR
# Discriminator menghasilkan dua output:
#   1. validitas real/fake
#   2. prediksi kelas Label 0/1
# ============================================================

def build_discriminator(feature_dim, num_classes):
    feature_input = Input(shape=(feature_dim,), name="feature_input")

    x = Dense(128)(feature_input)
    x = LeakyReLU(negative_slope=0.2)(x)
    x = Dropout(0.3)(x)

    x = Dense(64)(x)
    x = LeakyReLU(negative_slope=0.2)(x)
    x = Dropout(0.3)(x)

    x = Dense(32)(x)
    x = LeakyReLU(negative_slope=0.2)(x)

    validity_output = Dense(1, activation="sigmoid", name="validity_output")(x)
    class_output = Dense(num_classes, activation="softmax", name="class_output")(x)

    model = Model(feature_input, [validity_output, class_output], name="Discriminator")
    return model

generator = build_generator(noise_dim, num_classes, feature_dim)
discriminator = build_discriminator(feature_dim, num_classes)

# Compile discriminator
discriminator.compile(
    optimizer=Adam(learning_rate=0.0002, beta_1=0.5),
    loss={
        "validity_output": "binary_crossentropy",
        "class_output": "sparse_categorical_crossentropy"
    },
    metrics={
        "validity_output": "accuracy",
        "class_output": "accuracy"
    }
)

# ============================================================
# 4. BUILD COMBINED ACGAN MODEL
# Generator dilatih untuk menipu discriminator dan menghasilkan kelas sesuai label
# ============================================================

discriminator.trainable = False

noise_input = Input(shape=(noise_dim,), name="combined_noise_input")
label_input = Input(shape=(1,), dtype="int32", name="combined_label_input")

generated_features = generator([noise_input, label_input])
validity_output, class_output = discriminator(generated_features)

combined = Model(
    [noise_input, label_input],
    [validity_output, class_output],
    name="ACGAN_Combined"
)

combined.compile(
    optimizer=Adam(learning_rate=0.0002, beta_1=0.5),
    loss={
        "Discriminator": "binary_crossentropy"
    }
)

# Re-compile combined dengan output loss eksplisit
combined.compile(
    optimizer=Adam(learning_rate=0.0002, beta_1=0.5),
    loss=["binary_crossentropy", "sparse_categorical_crossentropy"]
)

print("\n=== Generator Summary ===")
generator.summary()

print("\n=== Discriminator Summary ===")
discriminator.summary()

# ============================================================
# 5. TRAINING ACGAN
# ============================================================

epochs = 1500
batch_size = 64
sample_interval = 100

d_losses = []
g_losses = []

half_batch = batch_size // 2

print("\n=== Mulai Training ACGAN ===")

for epoch in range(1, epochs + 1):

    # --------------------------------------------------------
    # Train Discriminator
    # --------------------------------------------------------

    # Ambil real samples
    idx = np.random.randint(0, X_train_acgan.shape[0], half_batch)
    real_features = X_train_acgan[idx]
    real_labels = y_train_acgan[idx]

    # Label validitas untuk real dan fake
    valid = np.ones((half_batch, 1)) * 0.9   # label smoothing
    fake = np.zeros((half_batch, 1))

    # Generate fake samples
    noise = np.random.normal(0, 1, (half_batch, noise_dim))
    sampled_labels = np.random.randint(0, num_classes, half_batch).reshape(-1, 1)
    fake_features = generator.predict([noise, sampled_labels], verbose=0)

    # Pastikan discriminator trainable saat dilatih
    discriminator.trainable = True

    # Train discriminator pada real
    d_loss_real = discriminator.train_on_batch(
        real_features,
        [valid, real_labels]
    )

    # Train discriminator pada fake
    d_loss_fake = discriminator.train_on_batch(
        fake_features,
        [fake, sampled_labels.flatten()]
    )

    # Rata-rata loss discriminator
    d_loss = 0.5 * np.add(d_loss_real[0], d_loss_fake[0])

    # --------------------------------------------------------
    # Train Generator
    # --------------------------------------------------------

    noise = np.random.normal(0, 1, (batch_size, noise_dim))
    sampled_labels = np.random.randint(0, num_classes, batch_size).reshape(-1, 1)

    valid_for_generator = np.ones((batch_size, 1))

    discriminator.trainable = False

    g_loss = combined.train_on_batch(
        [noise, sampled_labels],
        [valid_for_generator, sampled_labels.flatten()]
    )

    d_losses.append(d_loss)
    g_losses.append(g_loss[0] if isinstance(g_loss, list) else g_loss)

    # --------------------------------------------------------
    # Print progress
    # --------------------------------------------------------
    if epoch % sample_interval == 0 or epoch == 1:
        print(
            f"Epoch {epoch:4d}/{epochs} | "
            f"D loss: {float(d_loss):.4f} | "
            f"G loss: {float(g_losses[-1]):.4f}"
        )

print("\nTraining ACGAN selesai.")

# ============================================================
# 6. GENERATE SYNTHETIC DATA
# ============================================================

# Rasio synthetic data.
# 1.0 berarti jumlah synthetic per class = jumlah real training per class.
# Jika ingin lebih ringan, ubah ke 0.5.
synthetic_ratio = 1.0

n_label_0 = np.sum(y_train_acgan == 0)
n_label_1 = np.sum(y_train_acgan == 1)

n_synth_0 = int(n_label_0 * synthetic_ratio)
n_synth_1 = int(n_label_1 * synthetic_ratio)

print("\n=== Jumlah Synthetic Data yang Akan Dibuat ===")
print(f"Synthetic Label 0: {n_synth_0}")
print(f"Synthetic Label 1: {n_synth_1}")

# Generate label 0
noise_0 = np.random.normal(0, 1, (n_synth_0, noise_dim))
labels_0 = np.zeros((n_synth_0, 1), dtype=int)
X_synth_0_scaled = generator.predict([noise_0, labels_0], verbose=0)
y_synth_0 = np.zeros(n_synth_0, dtype=int)

# Generate label 1
noise_1 = np.random.normal(0, 1, (n_synth_1, noise_dim))
labels_1 = np.ones((n_synth_1, 1), dtype=int)
X_synth_1_scaled = generator.predict([noise_1, labels_1], verbose=0)
y_synth_1 = np.ones(n_synth_1, dtype=int)

# Gabungkan synthetic label 0 dan 1
X_synth_scaled = np.vstack([X_synth_0_scaled, X_synth_1_scaled])
y_synth = np.concatenate([y_synth_0, y_synth_1])

print("\n=== Synthetic Data Generated ===")
print(f"X_synth_scaled shape : {X_synth_scaled.shape}")
print(f"y_synth shape        : {y_synth.shape}")

unique_synth, counts_synth = np.unique(y_synth, return_counts=True)
print("\nDistribusi label synthetic:")
for u, c in zip(unique_synth, counts_synth):
    print(f"Label {u}: {c} sampel")

# ============================================================
# 7. GABUNGKAN REAL TRAINING + SYNTHETIC TRAINING
# Validation dan test TIDAK disentuh
# ============================================================

X_train_aug_scaled = np.vstack([X_train_scaled, X_synth_scaled])
y_train_aug = np.concatenate([np.array(y_train), y_synth])

# Shuffle augmented training data
X_train_aug_scaled, y_train_aug = shuffle(
    X_train_aug_scaled,
    y_train_aug,
    random_state=42
)

print("\n=== Augmented Training Data ===")
print(f"X_train_aug_scaled shape : {X_train_aug_scaled.shape}")
print(f"y_train_aug shape        : {y_train_aug.shape}")

unique_aug, counts_aug = np.unique(y_train_aug, return_counts=True)
print("\nDistribusi label setelah augmentation:")
for u, c in zip(unique_aug, counts_aug):
    print(f"Label {u}: {c} sampel")

# ============================================================
# 8. SIAPKAN DATA AUGMENTED UNTUK SVM, RF, DAN CNN
# ============================================================

# Untuk SVM: gunakan scaled augmented data
X_train_aug_svm = X_train_aug_scaled

# Untuk CNN: reshape menjadi (samples, 8, 1)
X_train_aug_cnn = X_train_aug_scaled.reshape(-1, feature_dim, 1)

# Untuk RF: gunakan versi unscaled.
# Synthetic data dibuat dalam scaled space, jadi kita inverse transform dulu.
X_synth_unscaled = scaler.inverse_transform(X_synth_scaled)
X_train_aug_unscaled = np.vstack([X_train_unscaled, X_synth_unscaled])
y_train_aug_rf = np.concatenate([np.array(y_train), y_synth])

X_train_aug_unscaled, y_train_aug_rf = shuffle(
    X_train_aug_unscaled,
    y_train_aug_rf,
    random_state=42
)

print("\n=== Variabel Augmented yang Siap Dipakai ===")
print(f"X_train_aug_svm shape      : {X_train_aug_svm.shape}")
print(f"X_train_aug_cnn shape      : {X_train_aug_cnn.shape}")
print(f"X_train_aug_unscaled shape : {X_train_aug_unscaled.shape}")
print(f"y_train_aug shape          : {y_train_aug.shape}")
print(f"y_train_aug_rf shape       : {y_train_aug_rf.shape}")

print("\nValidation dan test set tetap sama:")
print(f"X_val_scaled shape   : {X_val_scaled.shape}")
print(f"X_test_scaled shape  : {X_test_scaled.shape}")
print(f"X_val_cnn shape      : {X_val_cnn.shape}")
print(f"X_test_cnn shape     : {X_test_cnn.shape}")
print(f"X_val_unscaled shape : {X_val_unscaled.shape}")
print(f"X_test_unscaled shape: {X_test_unscaled.shape}")

# ============================================================
# 9. VISUALISASI LOSS ACGAN
# ============================================================

plt.figure(figsize=(10, 4))
plt.plot(d_losses, label="Discriminator Loss")
plt.plot(g_losses, label="Generator Loss")
plt.title("ACGAN Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.show()

print("\nTahap 4 selesai.")
print("Gunakan variabel berikut untuk training ulang model:")
print("  SVM : X_train_aug_svm, y_train_aug")
print("  RF  : X_train_aug_unscaled, y_train_aug_rf")
print("  CNN : X_train_aug_cnn, y_train_aug")
```

```python
# ============================================================
# TAHAP 5: Train Ulang SVM, RF, CNN dengan ACGAN Augmented
# Versi eksperimen: synthetic data dipakai untuk train, validation, dan test
# ============================================================

import numpy as np
import pandas as pd
import tensorflow as tf
import time

from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.utils import shuffle
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, confusion_matrix, classification_report
)

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (
    Input, Conv1D, MaxPooling1D, Flatten, Dense,
    Dropout, BatchNormalization
)
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# ============================================================
# 1. FUNGSI UNTUK GENERATE SYNTHETIC DATA BERDASARKAN DISTRIBUSI LABEL
# ============================================================

def generate_synthetic_by_label_distribution(y_reference, generator, noise_dim):
    """
    Membuat synthetic data dengan jumlah label 0 dan label 1
    mengikuti distribusi label dari y_reference.
    """
    y_reference = np.array(y_reference).astype(int)

    n_label_0 = np.sum(y_reference == 0)
    n_label_1 = np.sum(y_reference == 1)

    # Generate synthetic label 0
    noise_0 = np.random.normal(0, 1, (n_label_0, noise_dim))
    labels_0 = np.zeros((n_label_0, 1), dtype=int)
    X_synth_0 = generator.predict([noise_0, labels_0], verbose=0)
    y_synth_0 = np.zeros(n_label_0, dtype=int)

    # Generate synthetic label 1
    noise_1 = np.random.normal(0, 1, (n_label_1, noise_dim))
    labels_1 = np.ones((n_label_1, 1), dtype=int)
    X_synth_1 = generator.predict([noise_1, labels_1], verbose=0)
    y_synth_1 = np.ones(n_label_1, dtype=int)

    X_synth = np.vstack([X_synth_0, X_synth_1])
    y_synth = np.concatenate([y_synth_0, y_synth_1])

    X_synth, y_synth = shuffle(X_synth, y_synth, random_state=42)

    return X_synth, y_synth


# ============================================================
# 2. BUAT SYNTHETIC DATA UNTUK TRAIN, VALIDATION, DAN TEST
# ============================================================

np.random.seed(42)
tf.random.set_seed(42)

# Train synthetic sudah pernah dibuat di Tahap 4:
# X_synth_scaled, y_synth
# Tapi untuk konsistensi, kita buat ulang berdasarkan distribusi y_train
X_synth_train_scaled, y_synth_train = generate_synthetic_by_label_distribution(
    y_train, generator, noise_dim
)

X_synth_val_scaled, y_synth_val = generate_synthetic_by_label_distribution(
    y_val, generator, noise_dim
)

X_synth_test_scaled, y_synth_test = generate_synthetic_by_label_distribution(
    y_test, generator, noise_dim
)

print("=== Synthetic Data per Split ===")
print(f"Train synthetic      : {X_synth_train_scaled.shape}, {y_synth_train.shape}")
print(f"Validation synthetic : {X_synth_val_scaled.shape}, {y_synth_val.shape}")
print(f"Test synthetic       : {X_synth_test_scaled.shape}, {y_synth_test.shape}")

print("\nDistribusi label synthetic train:")
print(pd.Series(y_synth_train).value_counts().sort_index())

print("\nDistribusi label synthetic validation:")
print(pd.Series(y_synth_val).value_counts().sort_index())

print("\nDistribusi label synthetic test:")
print(pd.Series(y_synth_test).value_counts().sort_index())


# ============================================================
# 3. GABUNGKAN REAL + SYNTHETIC UNTUK TRAIN, VALIDATION, DAN TEST
# ============================================================

# Scaled data untuk SVM dan CNN
X_train_mix_scaled = np.vstack([X_train_scaled, X_synth_train_scaled])
y_train_mix = np.concatenate([np.array(y_train), y_synth_train])

X_val_mix_scaled = np.vstack([X_val_scaled, X_synth_val_scaled])
y_val_mix = np.concatenate([np.array(y_val), y_synth_val])

X_test_mix_scaled = np.vstack([X_test_scaled, X_synth_test_scaled])
y_test_mix = np.concatenate([np.array(y_test), y_synth_test])

# Shuffle tiap split
X_train_mix_scaled, y_train_mix = shuffle(X_train_mix_scaled, y_train_mix, random_state=42)
X_val_mix_scaled, y_val_mix = shuffle(X_val_mix_scaled, y_val_mix, random_state=42)
X_test_mix_scaled, y_test_mix = shuffle(X_test_mix_scaled, y_test_mix, random_state=42)

print("\n=== Mixed Scaled Data, Real + Synthetic ===")
print(f"X_train_mix_scaled : {X_train_mix_scaled.shape}")
print(f"y_train_mix        : {y_train_mix.shape}")
print(f"X_val_mix_scaled   : {X_val_mix_scaled.shape}")
print(f"y_val_mix          : {y_val_mix.shape}")
print(f"X_test_mix_scaled  : {X_test_mix_scaled.shape}")
print(f"y_test_mix         : {y_test_mix.shape}")

# Unscaled data untuk Random Forest
X_synth_train_unscaled = scaler.inverse_transform(X_synth_train_scaled)
X_synth_val_unscaled = scaler.inverse_transform(X_synth_val_scaled)
X_synth_test_unscaled = scaler.inverse_transform(X_synth_test_scaled)

X_train_mix_unscaled = np.vstack([X_train_unscaled, X_synth_train_unscaled])
y_train_mix_rf = np.concatenate([np.array(y_train), y_synth_train])

X_val_mix_unscaled = np.vstack([X_val_unscaled, X_synth_val_unscaled])
y_val_mix_rf = np.concatenate([np.array(y_val), y_synth_val])

X_test_mix_unscaled = np.vstack([X_test_unscaled, X_synth_test_unscaled])
y_test_mix_rf = np.concatenate([np.array(y_test), y_synth_test])

X_train_mix_unscaled, y_train_mix_rf = shuffle(X_train_mix_unscaled, y_train_mix_rf, random_state=42)
X_val_mix_unscaled, y_val_mix_rf = shuffle(X_val_mix_unscaled, y_val_mix_rf, random_state=42)
X_test_mix_unscaled, y_test_mix_rf = shuffle(X_test_mix_unscaled, y_test_mix_rf, random_state=42)

# CNN reshape
X_train_mix_cnn = X_train_mix_scaled.reshape(-1, feature_dim, 1)
X_val_mix_cnn = X_val_mix_scaled.reshape(-1, feature_dim, 1)
X_test_mix_cnn = X_test_mix_scaled.reshape(-1, feature_dim, 1)

print("\n=== Data Siap untuk Model ===")
print(f"SVM train : {X_train_mix_scaled.shape}")
print(f"SVM val   : {X_val_mix_scaled.shape}")
print(f"SVM test  : {X_test_mix_scaled.shape}")

print(f"RF train  : {X_train_mix_unscaled.shape}")
print(f"RF val    : {X_val_mix_unscaled.shape}")
print(f"RF test   : {X_test_mix_unscaled.shape}")

print(f"CNN train : {X_train_mix_cnn.shape}")
print(f"CNN val   : {X_val_mix_cnn.shape}")
print(f"CNN test  : {X_test_mix_cnn.shape}")


# ============================================================
# 4. FUNGSI EVALUASI
# ============================================================

def evaluate_model(model_name, y_val_true, y_val_pred, y_test_true, y_test_pred):
    print("\n" + "="*70)
    print(f"HASIL {model_name} - VALIDATION SET, REAL + SYNTHETIC")
    print("="*70)
    print(f"Accuracy  : {accuracy_score(y_val_true, y_val_pred):.4f}")
    print(f"Precision : {precision_score(y_val_true, y_val_pred, average='weighted'):.4f}")
    print(f"Recall    : {recall_score(y_val_true, y_val_pred, average='weighted'):.4f}")
    print(f"F1-Score  : {f1_score(y_val_true, y_val_pred, average='weighted'):.4f}")
    print(f"\nConfusion Matrix:\n{confusion_matrix(y_val_true, y_val_pred)}")
    print(f"\n{classification_report(y_val_true, y_val_pred, target_names=['Label 0 (Before)', 'Label 1 (After)'])}")

    print("="*70)
    print(f"HASIL {model_name} - TEST SET, REAL + SYNTHETIC")
    print("="*70)
    print(f"Accuracy  : {accuracy_score(y_test_true, y_test_pred):.4f}")
    print(f"Precision : {precision_score(y_test_true, y_test_pred, average='weighted'):.4f}")
    print(f"Recall    : {recall_score(y_test_true, y_test_pred, average='weighted'):.4f}")
    print(f"F1-Score  : {f1_score(y_test_true, y_test_pred, average='weighted'):.4f}")
    print(f"\nConfusion Matrix:\n{confusion_matrix(y_test_true, y_test_pred)}")
    print(f"\n{classification_report(y_test_true, y_test_pred, target_names=['Label 0 (Before)', 'Label 1 (After)'])}")


# ============================================================
# 5. TRAIN SVM DENGAN REAL + SYNTHETIC TRAIN
# EVALUASI PADA REAL + SYNTHETIC VALIDATION/TEST
# ============================================================

print("\n\nTraining SVM dengan train real + synthetic...")
start = time.time()

svm_mix_model = SVC(
    kernel='rbf',
    C=1.0,
    gamma='scale',
    random_state=42
)

svm_mix_model.fit(X_train_mix_scaled, y_train_mix)

svm_train_time = time.time() - start
print(f"Training SVM selesai dalam {svm_train_time:.2f} detik")

y_pred_val_svm_mix = svm_mix_model.predict(X_val_mix_scaled)
y_pred_test_svm_mix = svm_mix_model.predict(X_test_mix_scaled)

evaluate_model(
    "SVM + ACGAN MIXED",
    y_val_mix, y_pred_val_svm_mix,
    y_test_mix, y_pred_test_svm_mix
)


# ============================================================
# 6. TRAIN RANDOM FOREST DENGAN REAL + SYNTHETIC TRAIN
# EVALUASI PADA REAL + SYNTHETIC VALIDATION/TEST
# ============================================================

print("\n\nTraining Random Forest dengan train real + synthetic...")
start = time.time()

rf_mix_model = RandomForestClassifier(
    n_estimators=200,
    max_depth=None,
    min_samples_split=2,
    min_samples_leaf=1,
    random_state=42,
    n_jobs=-1
)

rf_mix_model.fit(X_train_mix_unscaled, y_train_mix_rf)

rf_train_time = time.time() - start
print(f"Training Random Forest selesai dalam {rf_train_time:.2f} detik")

y_pred_val_rf_mix = rf_mix_model.predict(X_val_mix_unscaled)
y_pred_test_rf_mix = rf_mix_model.predict(X_test_mix_unscaled)

evaluate_model(
    "RANDOM FOREST + ACGAN MIXED",
    y_val_mix_rf, y_pred_val_rf_mix,
    y_test_mix_rf, y_pred_test_rf_mix
)

print("\nFeature Importance RF + ACGAN Mixed:")
feat_imp_rf_mix = sorted(
    zip(gamma_cols, rf_mix_model.feature_importances_),
    key=lambda x: x[1],
    reverse=True
)

for feat, imp in feat_imp_rf_mix:
    print(f"{feat:30s} : {imp:.4f}")


# ============================================================
# 7. TRAIN CNN DENGAN REAL + SYNTHETIC TRAIN
# EVALUASI PADA REAL + SYNTHETIC VALIDATION/TEST
# ============================================================

print("\n\nTraining CNN dengan train real + synthetic...")

tf.random.set_seed(42)

model_cnn_mix = Sequential([
    Input(shape=(feature_dim, 1)),

    Conv1D(filters=32, kernel_size=3, activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling1D(pool_size=2),

    Conv1D(filters=64, kernel_size=3, activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling1D(pool_size=2),

    Flatten(),

    Dense(64, activation='relu'),
    Dropout(0.5),

    Dense(32, activation='relu'),
    Dropout(0.3),

    Dense(1, activation='sigmoid')
])

model_cnn_mix.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model_cnn_mix.summary()

early_stop_mix = EarlyStopping(
    monitor='val_loss',
    patience=15,
    restore_best_weights=True,
    verbose=1
)

reduce_lr_mix = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=5,
    min_lr=1e-6,
    verbose=1
)

start = time.time()

history_cnn_mix = model_cnn_mix.fit(
    X_train_mix_cnn, y_train_mix,
    validation_data=(X_val_mix_cnn, y_val_mix),
    epochs=100,
    batch_size=32,
    callbacks=[early_stop_mix, reduce_lr_mix],
    verbose=1
)

cnn_train_time = time.time() - start
print(f"\nTraining CNN selesai dalam {cnn_train_time:.2f} detik")
print(f"Total epoch berjalan: {len(history_cnn_mix.history['loss'])}")

y_prob_val_cnn_mix = model_cnn_mix.predict(X_val_mix_cnn, verbose=0).flatten()
y_prob_test_cnn_mix = model_cnn_mix.predict(X_test_mix_cnn, verbose=0).flatten()

y_pred_val_cnn_mix = (y_prob_val_cnn_mix >= 0.5).astype(int)
y_pred_test_cnn_mix = (y_prob_test_cnn_mix >= 0.5).astype(int)

evaluate_model(
    "CNN + ACGAN MIXED",
    y_val_mix, y_pred_val_cnn_mix,
    y_test_mix, y_pred_test_cnn_mix
)


# ============================================================
# 8. RINGKASAN HASIL MIXED, REAL + SYNTHETIC DI SEMUA SPLIT
# ============================================================

mixed_results = []

mixed_results.append({
    "Experiment": "ACGAN Mixed Train-Val-Test",
    "Model": "SVM",
    "Val Accuracy": accuracy_score(y_val_mix, y_pred_val_svm_mix),
    "Val Precision": precision_score(y_val_mix, y_pred_val_svm_mix, average='weighted'),
    "Val Recall": recall_score(y_val_mix, y_pred_val_svm_mix, average='weighted'),
    "Val F1": f1_score(y_val_mix, y_pred_val_svm_mix, average='weighted'),
    "Test Accuracy": accuracy_score(y_test_mix, y_pred_test_svm_mix),
    "Test Precision": precision_score(y_test_mix, y_pred_test_svm_mix, average='weighted'),
    "Test Recall": recall_score(y_test_mix, y_pred_test_svm_mix, average='weighted'),
    "Test F1": f1_score(y_test_mix, y_pred_test_svm_mix, average='weighted')
})

mixed_results.append({
    "Experiment": "ACGAN Mixed Train-Val-Test",
    "Model": "Random Forest",
    "Val Accuracy": accuracy_score(y_val_mix_rf, y_pred_val_rf_mix),
    "Val Precision": precision_score(y_val_mix_rf, y_pred_val_rf_mix, average='weighted'),
    "Val Recall": recall_score(y_val_mix_rf, y_pred_val_rf_mix, average='weighted'),
    "Val F1": f1_score(y_val_mix_rf, y_pred_val_rf_mix, average='weighted'),
    "Test Accuracy": accuracy_score(y_test_mix_rf, y_pred_test_rf_mix),
    "Test Precision": precision_score(y_test_mix_rf, y_pred_test_rf_mix, average='weighted'),
    "Test Recall": recall_score(y_test_mix_rf, y_pred_test_rf_mix, average='weighted'),
    "Test F1": f1_score(y_test_mix_rf, y_pred_test_rf_mix, average='weighted')
})

mixed_results.append({
    "Experiment": "ACGAN Mixed Train-Val-Test",
    "Model": "CNN",
    "Val Accuracy": accuracy_score(y_val_mix, y_pred_val_cnn_mix),
    "Val Precision": precision_score(y_val_mix, y_pred_val_cnn_mix, average='weighted'),
    "Val Recall": recall_score(y_val_mix, y_pred_val_cnn_mix, average='weighted'),
    "Val F1": f1_score(y_val_mix, y_pred_val_cnn_mix, average='weighted'),
    "Test Accuracy": accuracy_score(y_test_mix, y_pred_test_cnn_mix),
    "Test Precision": precision_score(y_test_mix, y_pred_test_cnn_mix, average='weighted'),
    "Test Recall": recall_score(y_test_mix, y_pred_test_cnn_mix, average='weighted'),
    "Test F1": f1_score(y_test_mix, y_pred_test_cnn_mix, average='weighted')
})

df_mixed_results = pd.DataFrame(mixed_results)

print("\n=== RINGKASAN HASIL ACGAN MIXED TRAIN-VALIDATION-TEST ===")
display(df_mixed_results.round(4))

df_mixed_results.to_csv("gamma_only_acgan_mixed_train_val_test_results.csv", index=False)

print("\nHasil disimpan sebagai gamma_only_acgan_mixed_train_val_test_results.csv")
```

```python
# ============================================================
# HYPERPARAMETER TUNING CNN
# ACGAN Mixed Train-Validation-Test
# Manual Search
# ============================================================

import numpy as np
import pandas as pd
import tensorflow as tf
import time

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (
    Input, Conv1D, MaxPooling1D, Flatten, Dense,
    Dropout, BatchNormalization
)
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, confusion_matrix, classification_report
)

# ============================================================
# 1. Function Build CNN
# ============================================================

def build_cnn_model(
    feature_dim=8,
    filters_1=32,
    filters_2=64,
    dense_units=64,
    dropout_1=0.5,
    dropout_2=0.3,
    learning_rate=0.001
):
    model = Sequential([
        Input(shape=(feature_dim, 1)),

        Conv1D(filters=filters_1, kernel_size=3, activation="relu", padding="same"),
        BatchNormalization(),
        MaxPooling1D(pool_size=2),

        Conv1D(filters=filters_2, kernel_size=3, activation="relu", padding="same"),
        BatchNormalization(),
        MaxPooling1D(pool_size=2),

        Flatten(),

        Dense(dense_units, activation="relu"),
        Dropout(dropout_1),

        Dense(dense_units // 2, activation="relu"),
        Dropout(dropout_2),

        Dense(1, activation="sigmoid")
    ])

    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate),
        loss="binary_crossentropy",
        metrics=["accuracy"]
    )

    return model


# ============================================================
# 2. Search Space CNN
# ============================================================

cnn_search_space = [
    {
        "filters_1": 32,
        "filters_2": 64,
        "dense_units": 64,
        "dropout_1": 0.4,
        "dropout_2": 0.2,
        "learning_rate": 0.001,
        "batch_size": 32
    },
    {
        "filters_1": 64,
        "filters_2": 128,
        "dense_units": 128,
        "dropout_1": 0.4,
        "dropout_2": 0.2,
        "learning_rate": 0.001,
        "batch_size": 32
    },
    {
        "filters_1": 64,
        "filters_2": 128,
        "dense_units": 128,
        "dropout_1": 0.3,
        "dropout_2": 0.2,
        "learning_rate": 0.0005,
        "batch_size": 32
    },
    {
        "filters_1": 128,
        "filters_2": 256,
        "dense_units": 256,
        "dropout_1": 0.3,
        "dropout_2": 0.2,
        "learning_rate": 0.0005,
        "batch_size": 64
    },
    {
        "filters_1": 128,
        "filters_2": 128,
        "dense_units": 128,
        "dropout_1": 0.2,
        "dropout_2": 0.1,
        "learning_rate": 0.0003,
        "batch_size": 64
    }
]

# ============================================================
# 3. Training beberapa kombinasi CNN
# ============================================================

cnn_tuning_results = []
best_val_acc = -1
best_cnn_tuned = None
best_cnn_params = None
best_cnn_history = None

print("Mulai tuning CNN...")

for i, params in enumerate(cnn_search_space, start=1):
    print("\n" + "="*80)
    print(f"Training CNN combination {i}/{len(cnn_search_space)}")
    print(params)
    print("="*80)

    tf.keras.backend.clear_session()
    tf.random.set_seed(42)
    np.random.seed(42)

    model = build_cnn_model(
        feature_dim=feature_dim,
        filters_1=params["filters_1"],
        filters_2=params["filters_2"],
        dense_units=params["dense_units"],
        dropout_1=params["dropout_1"],
        dropout_2=params["dropout_2"],
        learning_rate=params["learning_rate"]
    )

    early_stop = EarlyStopping(
        monitor="val_accuracy",
        patience=15,
        restore_best_weights=True,
        mode="max",
        verbose=1
    )

    reduce_lr = ReduceLROnPlateau(
        monitor="val_loss",
        factor=0.5,
        patience=5,
        min_lr=1e-6,
        verbose=1
    )

    start = time.time()

    history = model.fit(
        X_train_mix_cnn,
        y_train_mix,
        validation_data=(X_val_mix_cnn, y_val_mix),
        epochs=120,
        batch_size=params["batch_size"],
        callbacks=[early_stop, reduce_lr],
        verbose=1
    )

    train_time = time.time() - start

    val_loss, val_acc = model.evaluate(X_val_mix_cnn, y_val_mix, verbose=0)
    test_loss, test_acc = model.evaluate(X_test_mix_cnn, y_test_mix, verbose=0)

    cnn_tuning_results.append({
        "Combination": i,
        "filters_1": params["filters_1"],
        "filters_2": params["filters_2"],
        "dense_units": params["dense_units"],
        "dropout_1": params["dropout_1"],
        "dropout_2": params["dropout_2"],
        "learning_rate": params["learning_rate"],
        "batch_size": params["batch_size"],
        "Val Accuracy": val_acc,
        "Test Accuracy": test_acc,
        "Train Time": train_time,
        "Epochs": len(history.history["loss"])
    })

    print(f"\nCombination {i} selesai.")
    print(f"Val Accuracy  : {val_acc:.4f}")
    print(f"Test Accuracy : {test_acc:.4f}")

    if val_acc > best_val_acc:
        best_val_acc = val_acc
        best_cnn_tuned = model
        best_cnn_params = params
        best_cnn_history = history

print("\nTuning CNN selesai.")

df_cnn_tuning_results = pd.DataFrame(cnn_tuning_results)
display(df_cnn_tuning_results.round(4))

print("\nBest CNN Params:")
print(best_cnn_params)
print(f"Best Validation Accuracy: {best_val_acc:.4f}")

# ============================================================
# 4. Evaluasi Best CNN pada Validation dan Test Mixed
# ============================================================

y_prob_val_cnn_tuned = best_cnn_tuned.predict(X_val_mix_cnn, verbose=0).flatten()
y_prob_test_cnn_tuned = best_cnn_tuned.predict(X_test_mix_cnn, verbose=0).flatten()

y_pred_val_cnn_tuned = (y_prob_val_cnn_tuned >= 0.5).astype(int)
y_pred_test_cnn_tuned = (y_prob_test_cnn_tuned >= 0.5).astype(int)

print("\n" + "="*70)
print("HASIL TUNED CNN - VALIDATION SET MIXED")
print("="*70)
print(f"Accuracy  : {accuracy_score(y_val_mix, y_pred_val_cnn_tuned):.4f}")
print(f"Precision : {precision_score(y_val_mix, y_pred_val_cnn_tuned, average='weighted'):.4f}")
print(f"Recall    : {recall_score(y_val_mix, y_pred_val_cnn_tuned, average='weighted'):.4f}")
print(f"F1-Score  : {f1_score(y_val_mix, y_pred_val_cnn_tuned, average='weighted'):.4f}")
print(f"\nConfusion Matrix:\n{confusion_matrix(y_val_mix, y_pred_val_cnn_tuned)}")
print(f"\n{classification_report(y_val_mix, y_pred_val_cnn_tuned, target_names=['Label 0 (Before)', 'Label 1 (After)'])}")

print("="*70)
print("HASIL TUNED CNN - TEST SET MIXED")
print("="*70)
print(f"Accuracy  : {accuracy_score(y_test_mix, y_pred_test_cnn_tuned):.4f}")
print(f"Precision : {precision_score(y_test_mix, y_pred_test_cnn_tuned, average='weighted'):.4f}")
print(f"Recall    : {recall_score(y_test_mix, y_pred_test_cnn_tuned, average='weighted'):.4f}")
print(f"F1-Score  : {f1_score(y_test_mix, y_pred_test_cnn_tuned, average='weighted'):.4f}")
print(f"\nConfusion Matrix:\n{confusion_matrix(y_test_mix, y_pred_test_cnn_tuned)}")
print(f"\n{classification_report(y_test_mix, y_pred_test_cnn_tuned, target_names=['Label 0 (Before)', 'Label 1 (After)'])}")
```

```python
# ============================================================
# LABEL-AWARE TRAINING
# Real + Synthetic untuk Train, Validation, dan Test
# Fokus: model mengenali Label 0 dan Label 1 secara lebih seimbang
# ============================================================

import numpy as np
import pandas as pd
import tensorflow as tf
import time

from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier, VotingClassifier
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, classification_report, balanced_accuracy_score
)
from sklearn.utils.class_weight import compute_class_weight

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# ============================================================
# 1. Helper: cari threshold terbaik dari validation set
# ============================================================

def find_best_threshold(y_true, y_prob, metric="macro_f1"):
    """
    Mencari threshold terbaik untuk klasifikasi binary.
    Fokus agar Label 0 dan Label 1 lebih seimbang.
    """
    thresholds = np.arange(0.05, 0.96, 0.01)
    best_threshold = 0.5
    best_score = -1

    records = []

    for th in thresholds:
        y_pred = (y_prob >= th).astype(int)

        acc = accuracy_score(y_true, y_pred)
        bal_acc = balanced_accuracy_score(y_true, y_pred)
        macro_f1 = f1_score(y_true, y_pred, average="macro")
        weighted_f1 = f1_score(y_true, y_pred, average="weighted")

        records.append({
            "threshold": th,
            "accuracy": acc,
            "balanced_accuracy": bal_acc,
            "macro_f1": macro_f1,
            "weighted_f1": weighted_f1
        })

        if metric == "macro_f1":
            score = macro_f1
        elif metric == "balanced_accuracy":
            score = bal_acc
        else:
            score = acc

        if score > best_score:
            best_score = score
            best_threshold = th

    df_thresholds = pd.DataFrame(records)

    return best_threshold, best_score, df_thresholds


# ============================================================
# 2. Helper: evaluasi model dengan threshold pilihan
# ============================================================

def evaluate_binary_model(model_name, y_val_true, y_val_prob, y_test_true, y_test_prob, threshold):
    y_val_pred = (y_val_prob >= threshold).astype(int)
    y_test_pred = (y_test_prob >= threshold).astype(int)

    print("\n" + "="*80)
    print(f"{model_name} | BEST THRESHOLD = {threshold:.2f}")
    print("="*80)

    print("\nVALIDATION SET, REAL + SYNTHETIC")
    print("-"*80)
    print(f"Accuracy          : {accuracy_score(y_val_true, y_val_pred):.4f}")
    print(f"Balanced Accuracy : {balanced_accuracy_score(y_val_true, y_val_pred):.4f}")
    print(f"Precision Macro   : {precision_score(y_val_true, y_val_pred, average='macro'):.4f}")
    print(f"Recall Macro      : {recall_score(y_val_true, y_val_pred, average='macro'):.4f}")
    print(f"F1 Macro          : {f1_score(y_val_true, y_val_pred, average='macro'):.4f}")
    print(f"F1 Weighted       : {f1_score(y_val_true, y_val_pred, average='weighted'):.4f}")
    print(f"\nConfusion Matrix:\n{confusion_matrix(y_val_true, y_val_pred)}")
    print("\nClassification Report:")
    print(classification_report(
        y_val_true,
        y_val_pred,
        target_names=["Label 0 (Before)", "Label 1 (After)"]
    ))

    print("\nTEST SET, REAL + SYNTHETIC")
    print("-"*80)
    print(f"Accuracy          : {accuracy_score(y_test_true, y_test_pred):.4f}")
    print(f"Balanced Accuracy : {balanced_accuracy_score(y_test_true, y_test_pred):.4f}")
    print(f"Precision Macro   : {precision_score(y_test_true, y_test_pred, average='macro'):.4f}")
    print(f"Recall Macro      : {recall_score(y_test_true, y_test_pred, average='macro'):.4f}")
    print(f"F1 Macro          : {f1_score(y_test_true, y_test_pred, average='macro'):.4f}")
    print(f"F1 Weighted       : {f1_score(y_test_true, y_test_pred, average='weighted'):.4f}")
    print(f"\nConfusion Matrix:\n{confusion_matrix(y_test_true, y_test_pred)}")
    print("\nClassification Report:")
    print(classification_report(
        y_test_true,
        y_test_pred,
        target_names=["Label 0 (Before)", "Label 1 (After)"]
    ))

    return {
        "Model": model_name,
        "Best Threshold": threshold,
        "Val Accuracy": accuracy_score(y_val_true, y_val_pred),
        "Val Balanced Accuracy": balanced_accuracy_score(y_val_true, y_val_pred),
        "Val Macro F1": f1_score(y_val_true, y_val_pred, average="macro"),
        "Val Weighted F1": f1_score(y_val_true, y_val_pred, average="weighted"),
        "Test Accuracy": accuracy_score(y_test_true, y_test_pred),
        "Test Balanced Accuracy": balanced_accuracy_score(y_test_true, y_test_pred),
        "Test Macro F1": f1_score(y_test_true, y_test_pred, average="macro"),
        "Test Weighted F1": f1_score(y_test_true, y_test_pred, average="weighted")
    }


# ============================================================
# 3. Class weight untuk membantu model memperhatikan dua label
# ============================================================

classes = np.array([0, 1])
class_weights_array = compute_class_weight(
    class_weight="balanced",
    classes=classes,
    y=y_train_mix
)

class_weights_dict = {
    0: class_weights_array[0],
    1: class_weights_array[1]
}

print("Class weights:")
print(class_weights_dict)


# ============================================================
# 4. MODEL 1: SVM probability + threshold tuning
# ============================================================

print("\n\nTraining label-aware SVM...")
start = time.time()

svm_label_model = SVC(
    kernel="rbf",
    C=50,
    gamma=0.1,
    probability=True,
    class_weight="balanced",
    random_state=42
)

svm_label_model.fit(X_train_mix_scaled, y_train_mix)

print(f"SVM training time: {time.time() - start:.2f} detik")

y_val_prob_svm_label = svm_label_model.predict_proba(X_val_mix_scaled)[:, 1]
y_test_prob_svm_label = svm_label_model.predict_proba(X_test_mix_scaled)[:, 1]

best_th_svm, best_score_svm, df_th_svm = find_best_threshold(
    y_val_mix,
    y_val_prob_svm_label,
    metric="macro_f1"
)

result_svm_label = evaluate_binary_model(
    "SVM Label-Aware",
    y_val_mix,
    y_val_prob_svm_label,
    y_test_mix,
    y_test_prob_svm_label,
    best_th_svm
)


# ============================================================
# 5. MODEL 2: Random Forest lebih kuat + threshold tuning
# ============================================================

print("\n\nTraining label-aware Random Forest...")
start = time.time()

rf_label_model = RandomForestClassifier(
    n_estimators=1000,
    max_depth=None,
    min_samples_split=2,
    min_samples_leaf=1,
    max_features="sqrt",
    bootstrap=False,
    class_weight="balanced_subsample",
    random_state=42,
    n_jobs=-1
)

rf_label_model.fit(X_train_mix_unscaled, y_train_mix_rf)

print(f"Random Forest training time: {time.time() - start:.2f} detik")

y_val_prob_rf_label = rf_label_model.predict_proba(X_val_mix_unscaled)[:, 1]
y_test_prob_rf_label = rf_label_model.predict_proba(X_test_mix_unscaled)[:, 1]

best_th_rf, best_score_rf, df_th_rf = find_best_threshold(
    y_val_mix_rf,
    y_val_prob_rf_label,
    metric="macro_f1"
)

result_rf_label = evaluate_binary_model(
    "Random Forest Label-Aware",
    y_val_mix_rf,
    y_val_prob_rf_label,
    y_test_mix_rf,
    y_test_prob_rf_label,
    best_th_rf
)


# ============================================================
# 6. MODEL 3: Extra Trees, sering lebih kuat untuk tabular features
# ============================================================

print("\n\nTraining label-aware Extra Trees...")
start = time.time()

et_label_model = ExtraTreesClassifier(
    n_estimators=1000,
    max_depth=None,
    min_samples_split=2,
    min_samples_leaf=1,
    max_features="sqrt",
    bootstrap=False,
    class_weight="balanced",
    random_state=42,
    n_jobs=-1
)

et_label_model.fit(X_train_mix_unscaled, y_train_mix_rf)

print(f"Extra Trees training time: {time.time() - start:.2f} detik")

y_val_prob_et_label = et_label_model.predict_proba(X_val_mix_unscaled)[:, 1]
y_test_prob_et_label = et_label_model.predict_proba(X_test_mix_unscaled)[:, 1]

best_th_et, best_score_et, df_th_et = find_best_threshold(
    y_val_mix_rf,
    y_val_prob_et_label,
    metric="macro_f1"
)

result_et_label = evaluate_binary_model(
    "Extra Trees Label-Aware",
    y_val_mix_rf,
    y_val_prob_et_label,
    y_test_mix_rf,
    y_test_prob_et_label,
    best_th_et
)


# ============================================================
# 7. MODEL 4: MLP tabular, lebih cocok daripada CNN untuk 8 fitur tabular
# ============================================================

print("\n\nTraining label-aware MLP untuk tabular Gamma features...")

tf.keras.backend.clear_session()
tf.random.set_seed(42)
np.random.seed(42)

mlp_label_model = Sequential([
    Input(shape=(feature_dim,)),

    Dense(128, activation="relu"),
    BatchNormalization(),
    Dropout(0.25),

    Dense(128, activation="relu"),
    BatchNormalization(),
    Dropout(0.20),

    Dense(64, activation="relu"),
    BatchNormalization(),
    Dropout(0.15),

    Dense(32, activation="relu"),
    Dropout(0.10),

    Dense(1, activation="sigmoid")
])

mlp_label_model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.0005),
    loss="binary_crossentropy",
    metrics=["accuracy"]
)

early_stop_mlp = EarlyStopping(
    monitor="val_loss",
    patience=20,
    restore_best_weights=True,
    verbose=1
)

reduce_lr_mlp = ReduceLROnPlateau(
    monitor="val_loss",
    factor=0.5,
    patience=7,
    min_lr=1e-6,
    verbose=1
)

start = time.time()

history_mlp_label = mlp_label_model.fit(
    X_train_mix_scaled,
    y_train_mix,
    validation_data=(X_val_mix_scaled, y_val_mix),
    epochs=200,
    batch_size=32,
    class_weight=class_weights_dict,
    callbacks=[early_stop_mlp, reduce_lr_mlp],
    verbose=1
)

print(f"MLP training time: {time.time() - start:.2f} detik")

y_val_prob_mlp_label = mlp_label_model.predict(X_val_mix_scaled, verbose=0).flatten()
y_test_prob_mlp_label = mlp_label_model.predict(X_test_mix_scaled, verbose=0).flatten()

best_th_mlp, best_score_mlp, df_th_mlp = find_best_threshold(
    y_val_mix,
    y_val_prob_mlp_label,
    metric="macro_f1"
)

result_mlp_label = evaluate_binary_model(
    "MLP Label-Aware",
    y_val_mix,
    y_val_prob_mlp_label,
    y_test_mix,
    y_test_prob_mlp_label,
    best_th_mlp
)


# ============================================================
# 8. Ensemble probabilitas: RF + Extra Trees + SVM + MLP
# ============================================================

print("\n\nEvaluasi Ensemble Probability Averaging...")

# Pastikan semua probabilitas dari validation/test memakai urutan data yang sama
y_val_prob_ensemble = (
    y_val_prob_svm_label +
    y_val_prob_rf_label +
    y_val_prob_et_label +
    y_val_prob_mlp_label
) / 4

y_test_prob_ensemble = (
    y_test_prob_svm_label +
    y_test_prob_rf_label +
    y_test_prob_et_label +
    y_test_prob_mlp_label
) / 4

best_th_ensemble, best_score_ensemble, df_th_ensemble = find_best_threshold(
    y_val_mix,
    y_val_prob_ensemble,
    metric="macro_f1"
)

result_ensemble_label = evaluate_binary_model(
    "Ensemble Label-Aware",
    y_val_mix,
    y_val_prob_ensemble,
    y_test_mix,
    y_test_prob_ensemble,
    best_th_ensemble
)


# ============================================================
# 9. Ringkasan semua model label-aware
# ============================================================

label_aware_results = pd.DataFrame([
    result_svm_label,
    result_rf_label,
    result_et_label,
    result_mlp_label,
    result_ensemble_label
])

print("\n" + "="*80)
print("RINGKASAN HASIL LABEL-AWARE MODELS")
print("="*80)

display(label_aware_results.sort_values(
    by="Test Macro F1",
    ascending=False
).round(4))

label_aware_results.to_csv(
    "gamma_only_acgan_mixed_label_aware_results.csv",
    index=False
)

print("\nHasil disimpan sebagai gamma_only_acgan_mixed_label_aware_results.csv")


# ============================================================
# 10. Lihat threshold terbaik setiap model
# ============================================================

print("\nBest threshold per model:")
print(f"SVM        : {best_th_svm:.2f}")
print(f"RF         : {best_th_rf:.2f}")
print(f"ExtraTrees : {best_th_et:.2f}")
print(f"MLP        : {best_th_mlp:.2f}")
print(f"Ensemble   : {best_th_ensemble:.2f}")
```

```python
# ============================================================
# TAHAP 5 REVISI:
# LABEL-CONSISTENT ACGAN MIXED TRAIN-VALIDATION-TEST
# Fokus: SVM, Random Forest, CNN mampu mengenali Label 0 dan Label 1
# ============================================================

import numpy as np
import pandas as pd
import tensorflow as tf
import time

from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.utils import shuffle
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, classification_report, balanced_accuracy_score
)

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (
    Input, Conv1D, MaxPooling1D, Flatten, Dense,
    Dropout, BatchNormalization
)
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

np.random.seed(42)
tf.random.set_seed(42)

# ============================================================
# 1. Fungsi cek label synthetic menggunakan discriminator ACGAN
# ============================================================

def discriminator_class_confidence(discriminator, X_data):
    """
    Mengambil prediksi class dari discriminator ACGAN.
    Output discriminator: [validity_output, class_output]
    """
    pred = discriminator.predict(X_data, verbose=0)
    class_probs = pred[1]
    pred_labels = np.argmax(class_probs, axis=1)
    pred_conf = np.max(class_probs, axis=1)
    return pred_labels, pred_conf, class_probs


def generate_label_consistent_synthetic(
    target_label,
    n_target,
    generator,
    discriminator,
    noise_dim,
    confidence_threshold=0.70,
    max_rounds=30,
    batch_multiplier=4
):
    """
    Generate synthetic data untuk satu label.
    Hanya sample yang oleh discriminator diprediksi sebagai label target
    dan confidence-nya cukup tinggi yang disimpan.
    """

    collected_X = []
    collected_y = []

    total_collected = 0
    round_id = 0

    while total_collected < n_target and round_id < max_rounds:
        round_id += 1

        n_generate = max((n_target - total_collected) * batch_multiplier, 256)

        noise = np.random.normal(0, 1, (n_generate, noise_dim))
        labels = np.full((n_generate, 1), target_label, dtype=int)

        X_gen = generator.predict([noise, labels], verbose=0)

        pred_labels, pred_conf, _ = discriminator_class_confidence(
            discriminator,
            X_gen
        )

        keep_mask = (pred_labels == target_label) & (pred_conf >= confidence_threshold)

        X_keep = X_gen[keep_mask]
        y_keep = np.full(X_keep.shape[0], target_label, dtype=int)

        if X_keep.shape[0] > 0:
            collected_X.append(X_keep)
            collected_y.append(y_keep)
            total_collected += X_keep.shape[0]

        print(
            f"Label {target_label} | round {round_id:02d} | "
            f"generated {n_generate} | kept {X_keep.shape[0]} | "
            f"total kept {total_collected}/{n_target}"
        )

    if total_collected == 0:
        raise ValueError(f"Tidak ada synthetic data yang lolos untuk Label {target_label}.")

    X_all = np.vstack(collected_X)
    y_all = np.concatenate(collected_y)

    # Ambil tepat sebanyak target
    X_all = X_all[:n_target]
    y_all = y_all[:n_target]

    return X_all, y_all


def generate_split_synthetic_label_consistent(
    y_reference,
    generator,
    discriminator,
    noise_dim,
    confidence_threshold=0.70
):
    """
    Generate synthetic data untuk satu split mengikuti distribusi label y_reference.
    """

    y_reference = np.array(y_reference).astype(int)

    n0 = np.sum(y_reference == 0)
    n1 = np.sum(y_reference == 1)

    print("\n" + "="*80)
    print(f"Generate label-consistent synthetic split | Label 0: {n0}, Label 1: {n1}")
    print("="*80)

    X0, y0 = generate_label_consistent_synthetic(
        target_label=0,
        n_target=n0,
        generator=generator,
        discriminator=discriminator,
        noise_dim=noise_dim,
        confidence_threshold=confidence_threshold
    )

    X1, y1 = generate_label_consistent_synthetic(
        target_label=1,
        n_target=n1,
        generator=generator,
        discriminator=discriminator,
        noise_dim=noise_dim,
        confidence_threshold=confidence_threshold
    )

    X_synth = np.vstack([X0, X1])
    y_synth = np.concatenate([y0, y1])

    X_synth, y_synth = shuffle(X_synth, y_synth, random_state=42)

    return X_synth, y_synth


# ============================================================
# 2. Generate synthetic yang label-consistent untuk train, val, test
# ============================================================

confidence_threshold = 0.70

X_synth_train_lc, y_synth_train_lc = generate_split_synthetic_label_consistent(
    y_train,
    generator,
    discriminator,
    noise_dim,
    confidence_threshold=confidence_threshold
)

X_synth_val_lc, y_synth_val_lc = generate_split_synthetic_label_consistent(
    y_val,
    generator,
    discriminator,
    noise_dim,
    confidence_threshold=confidence_threshold
)

X_synth_test_lc, y_synth_test_lc = generate_split_synthetic_label_consistent(
    y_test,
    generator,
    discriminator,
    noise_dim,
    confidence_threshold=confidence_threshold
)

print("\n=== Synthetic Label-Consistent Data ===")
print("Train synthetic:", X_synth_train_lc.shape, y_synth_train_lc.shape)
print("Val synthetic  :", X_synth_val_lc.shape, y_synth_val_lc.shape)
print("Test synthetic :", X_synth_test_lc.shape, y_synth_test_lc.shape)

print("\nDistribusi synthetic train:")
print(pd.Series(y_synth_train_lc).value_counts().sort_index())

print("\nDistribusi synthetic validation:")
print(pd.Series(y_synth_val_lc).value_counts().sort_index())

print("\nDistribusi synthetic test:")
print(pd.Series(y_synth_test_lc).value_counts().sort_index())


# ============================================================
# 3. Gabungkan real + synthetic untuk train, validation, test
# ============================================================

# Untuk SVM dan CNN, pakai scaled data
X_train_lc_scaled = np.vstack([X_train_scaled, X_synth_train_lc])
y_train_lc = np.concatenate([np.array(y_train), y_synth_train_lc])

X_val_lc_scaled = np.vstack([X_val_scaled, X_synth_val_lc])
y_val_lc = np.concatenate([np.array(y_val), y_synth_val_lc])

X_test_lc_scaled = np.vstack([X_test_scaled, X_synth_test_lc])
y_test_lc = np.concatenate([np.array(y_test), y_synth_test_lc])

X_train_lc_scaled, y_train_lc = shuffle(X_train_lc_scaled, y_train_lc, random_state=42)
X_val_lc_scaled, y_val_lc = shuffle(X_val_lc_scaled, y_val_lc, random_state=42)
X_test_lc_scaled, y_test_lc = shuffle(X_test_lc_scaled, y_test_lc, random_state=42)

# Untuk RF, inverse transform ke unscaled
X_synth_train_lc_unscaled = scaler.inverse_transform(X_synth_train_lc)
X_synth_val_lc_unscaled = scaler.inverse_transform(X_synth_val_lc)
X_synth_test_lc_unscaled = scaler.inverse_transform(X_synth_test_lc)

X_train_lc_unscaled = np.vstack([X_train_unscaled, X_synth_train_lc_unscaled])
y_train_lc_rf = np.concatenate([np.array(y_train), y_synth_train_lc])

X_val_lc_unscaled = np.vstack([X_val_unscaled, X_synth_val_lc_unscaled])
y_val_lc_rf = np.concatenate([np.array(y_val), y_synth_val_lc])

X_test_lc_unscaled = np.vstack([X_test_unscaled, X_synth_test_lc_unscaled])
y_test_lc_rf = np.concatenate([np.array(y_test), y_synth_test_lc])

X_train_lc_unscaled, y_train_lc_rf = shuffle(X_train_lc_unscaled, y_train_lc_rf, random_state=42)
X_val_lc_unscaled, y_val_lc_rf = shuffle(X_val_lc_unscaled, y_val_lc_rf, random_state=42)
X_test_lc_unscaled, y_test_lc_rf = shuffle(X_test_lc_unscaled, y_test_lc_rf, random_state=42)

# CNN shape
X_train_lc_cnn = X_train_lc_scaled.reshape(-1, feature_dim, 1)
X_val_lc_cnn = X_val_lc_scaled.reshape(-1, feature_dim, 1)
X_test_lc_cnn = X_test_lc_scaled.reshape(-1, feature_dim, 1)

print("\n=== Mixed Label-Consistent Dataset ===")
print("SVM train:", X_train_lc_scaled.shape, y_train_lc.shape)
print("SVM val  :", X_val_lc_scaled.shape, y_val_lc.shape)
print("SVM test :", X_test_lc_scaled.shape, y_test_lc.shape)

print("RF train :", X_train_lc_unscaled.shape, y_train_lc_rf.shape)
print("RF val   :", X_val_lc_unscaled.shape, y_val_lc_rf.shape)
print("RF test  :", X_test_lc_unscaled.shape, y_test_lc_rf.shape)

print("CNN train:", X_train_lc_cnn.shape)
print("CNN val  :", X_val_lc_cnn.shape)
print("CNN test :", X_test_lc_cnn.shape)


# ============================================================
# 4. Threshold khusus untuk mengenali dua label secara seimbang
# ============================================================

def find_threshold_balanced_labels(y_true, y_prob):
    """
    Cari threshold yang membuat recall Label 0 dan Label 1 seimbang.
    Score utama = min(recall_0, recall_1)
    Tie-breaker = accuracy.
    """

    thresholds = np.arange(0.05, 0.96, 0.01)

    best_threshold = 0.5
    best_score = -1
    best_acc = -1
    records = []

    for th in thresholds:
        y_pred = (y_prob >= th).astype(int)

        cm = confusion_matrix(y_true, y_pred, labels=[0, 1])
        tn, fp, fn, tp = cm.ravel()

        recall_0 = tn / (tn + fp + 1e-9)
        recall_1 = tp / (tp + fn + 1e-9)

        min_recall = min(recall_0, recall_1)
        acc = accuracy_score(y_true, y_pred)
        bal_acc = balanced_accuracy_score(y_true, y_pred)
        macro_f1 = f1_score(y_true, y_pred, average="macro")

        records.append({
            "threshold": th,
            "recall_label_0": recall_0,
            "recall_label_1": recall_1,
            "min_recall": min_recall,
            "accuracy": acc,
            "balanced_accuracy": bal_acc,
            "macro_f1": macro_f1
        })

        if (min_recall > best_score) or (min_recall == best_score and acc > best_acc):
            best_score = min_recall
            best_acc = acc
            best_threshold = th

    df_thresholds = pd.DataFrame(records)

    return best_threshold, df_thresholds


def evaluate_with_balanced_threshold(model_name, y_val_true, y_val_prob, y_test_true, y_test_prob):
    best_th, df_thresholds = find_threshold_balanced_labels(y_val_true, y_val_prob)

    y_val_pred = (y_val_prob >= best_th).astype(int)
    y_test_pred = (y_test_prob >= best_th).astype(int)

    print("\n" + "="*90)
    print(f"{model_name} | THRESHOLD SEIMBANG = {best_th:.2f}")
    print("="*90)

    print("\nVALIDATION SET")
    print("-"*90)
    print(f"Accuracy          : {accuracy_score(y_val_true, y_val_pred):.4f}")
    print(f"Balanced Accuracy : {balanced_accuracy_score(y_val_true, y_val_pred):.4f}")
    print(f"Macro F1          : {f1_score(y_val_true, y_val_pred, average='macro'):.4f}")
    print(f"Weighted F1       : {f1_score(y_val_true, y_val_pred, average='weighted'):.4f}")
    print(f"\nConfusion Matrix:\n{confusion_matrix(y_val_true, y_val_pred)}")
    print(classification_report(
        y_val_true,
        y_val_pred,
        target_names=["Label 0 (Before)", "Label 1 (After)"]
    ))

    print("\nTEST SET")
    print("-"*90)
    print(f"Accuracy          : {accuracy_score(y_test_true, y_test_pred):.4f}")
    print(f"Balanced Accuracy : {balanced_accuracy_score(y_test_true, y_test_pred):.4f}")
    print(f"Macro F1          : {f1_score(y_test_true, y_test_pred, average='macro'):.4f}")
    print(f"Weighted F1       : {f1_score(y_test_true, y_test_pred, average='weighted'):.4f}")
    print(f"\nConfusion Matrix:\n{confusion_matrix(y_test_true, y_test_pred)}")
    print(classification_report(
        y_test_true,
        y_test_pred,
        target_names=["Label 0 (Before)", "Label 1 (After)"]
    ))

    return {
        "Model": model_name,
        "Best Threshold": best_th,
        "Val Accuracy": accuracy_score(y_val_true, y_val_pred),
        "Val Balanced Accuracy": balanced_accuracy_score(y_val_true, y_val_pred),
        "Val Macro F1": f1_score(y_val_true, y_val_pred, average="macro"),
        "Val Weighted F1": f1_score(y_val_true, y_val_pred, average="weighted"),
        "Test Accuracy": accuracy_score(y_test_true, y_test_pred),
        "Test Balanced Accuracy": balanced_accuracy_score(y_test_true, y_test_pred),
        "Test Macro F1": f1_score(y_test_true, y_test_pred, average="macro"),
        "Test Weighted F1": f1_score(y_test_true, y_test_pred, average="weighted")
    }, df_thresholds


# ============================================================
# 5. Train SVM
# ============================================================

print("\n\nTraining SVM pada label-consistent mixed data...")
start = time.time()

svm_lc_model = SVC(
    kernel="rbf",
    C=100,
    gamma=0.05,
    probability=True,
    class_weight="balanced",
    random_state=42
)

svm_lc_model.fit(X_train_lc_scaled, y_train_lc)

print(f"SVM training selesai: {time.time() - start:.2f} detik")

y_val_prob_svm_lc = svm_lc_model.predict_proba(X_val_lc_scaled)[:, 1]
y_test_prob_svm_lc = svm_lc_model.predict_proba(X_test_lc_scaled)[:, 1]

result_svm_lc, df_th_svm_lc = evaluate_with_balanced_threshold(
    "SVM Label-Consistent ACGAN",
    y_val_lc,
    y_val_prob_svm_lc,
    y_test_lc,
    y_test_prob_svm_lc
)


# ============================================================
# 6. Train Random Forest
# ============================================================

print("\n\nTraining Random Forest pada label-consistent mixed data...")
start = time.time()

rf_lc_model = RandomForestClassifier(
    n_estimators=1200,
    max_depth=None,
    min_samples_split=2,
    min_samples_leaf=1,
    max_features=None,
    bootstrap=False,
    class_weight="balanced",
    random_state=42,
    n_jobs=-1
)

rf_lc_model.fit(X_train_lc_unscaled, y_train_lc_rf)

print(f"Random Forest training selesai: {time.time() - start:.2f} detik")

y_val_prob_rf_lc = rf_lc_model.predict_proba(X_val_lc_unscaled)[:, 1]
y_test_prob_rf_lc = rf_lc_model.predict_proba(X_test_lc_unscaled)[:, 1]

result_rf_lc, df_th_rf_lc = evaluate_with_balanced_threshold(
    "Random Forest Label-Consistent ACGAN",
    y_val_lc_rf,
    y_val_prob_rf_lc,
    y_test_lc_rf,
    y_test_prob_rf_lc
)

print("\nFeature Importance RF Label-Consistent:")
rf_lc_importance = sorted(
    zip(gamma_cols, rf_lc_model.feature_importances_),
    key=lambda x: x[1],
    reverse=True
)
for feat, imp in rf_lc_importance:
    print(f"{feat:30s} : {imp:.4f}")


# ============================================================
# 7. Train CNN
# ============================================================

print("\n\nTraining CNN pada label-consistent mixed data...")

tf.keras.backend.clear_session()
tf.random.set_seed(42)
np.random.seed(42)

cnn_lc_model = Sequential([
    Input(shape=(feature_dim, 1)),

    Conv1D(64, kernel_size=3, activation="relu", padding="same"),
    BatchNormalization(),
    MaxPooling1D(pool_size=2),
    Dropout(0.20),

    Conv1D(128, kernel_size=3, activation="relu", padding="same"),
    BatchNormalization(),
    MaxPooling1D(pool_size=2),
    Dropout(0.20),

    Flatten(),

    Dense(128, activation="relu"),
    BatchNormalization(),
    Dropout(0.30),

    Dense(64, activation="relu"),
    BatchNormalization(),
    Dropout(0.20),

    Dense(1, activation="sigmoid")
])

cnn_lc_model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.0005),
    loss="binary_crossentropy",
    metrics=["accuracy"]
)

early_stop_cnn_lc = EarlyStopping(
    monitor="val_loss",
    patience=20,
    restore_best_weights=True,
    verbose=1
)

reduce_lr_cnn_lc = ReduceLROnPlateau(
    monitor="val_loss",
    factor=0.5,
    patience=7,
    min_lr=1e-6,
    verbose=1
)

start = time.time()

history_cnn_lc = cnn_lc_model.fit(
    X_train_lc_cnn,
    y_train_lc,
    validation_data=(X_val_lc_cnn, y_val_lc),
    epochs=200,
    batch_size=32,
    callbacks=[early_stop_cnn_lc, reduce_lr_cnn_lc],
    verbose=1
)

print(f"CNN training selesai: {time.time() - start:.2f} detik")

y_val_prob_cnn_lc = cnn_lc_model.predict(X_val_lc_cnn, verbose=0).flatten()
y_test_prob_cnn_lc = cnn_lc_model.predict(X_test_lc_cnn, verbose=0).flatten()

result_cnn_lc, df_th_cnn_lc = evaluate_with_balanced_threshold(
    "CNN Label-Consistent ACGAN",
    y_val_lc,
    y_val_prob_cnn_lc,
    y_test_lc,
    y_test_prob_cnn_lc
)


# ============================================================
# 8. Ringkasan akhir
# ============================================================

df_label_consistent_results = pd.DataFrame([
    result_svm_lc,
    result_rf_lc,
    result_cnn_lc
])

print("\n" + "="*90)
print("RINGKASAN LABEL-CONSISTENT ACGAN MIXED RESULTS")
print("="*90)

display(
    df_label_consistent_results
    .sort_values(by="Test Balanced Accuracy", ascending=False)
    .round(4)
)

df_label_consistent_results.to_csv(
    "gamma_only_acgan_label_consistent_svm_rf_cnn_results.csv",
    index=False
)

print("\nHasil disimpan sebagai gamma_only_acgan_label_consistent_svm_rf_cnn_results.csv")
```
